## Handle Missing Values

```python
# Fill missing 'Size' values within each group based on 'Product Name' using median
df['Size'] = df.groupby('Product Name')['Size'].transform(lambda x: x.fillna(x.median()))
```

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data Overview
print(df.shape)
print(df.dtypes)
print(df.head())

# Descriptive Statistics
print(df.describe())

# Univariate Analysis: Histograms
df['Price'].hist(bins=30)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()



# Bivariate Analysis: Scatter Plot
sns.scatterplot(data=df, x='Price', y='Size')
plt.title('Price vs Size')
plt.xlabel('Price')
plt.ylabel('Size')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame

# Step 1: Filter to keep only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Step 2: Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Step 3: Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
df['Message Date'] = pd.to_datetime(df['Message Date'], format='%m/%d/%Y %H:%M')
df['Month'] = df['Message Date'].dt.month
df['Year'] = df['Message Date'].dt.year
```

```python
price_trend = df.groupby('Message Date')['Price'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Message Date', y='Price', data=price_trend)
plt.title('Average Price Over Time')
plt.xticks(rotation=45)
plt.show()
```
